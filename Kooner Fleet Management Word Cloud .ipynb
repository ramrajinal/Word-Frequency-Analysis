# **Word Frequency Analysis of Google Reviews and Indeed**

### **Purpose**
To perform a simple Word Frequency Analysis using Google Reviews and Indeed to gain insight into the sentiment surrounding Kooner Fleet Management.

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#Load csv file using Pandas
df=r"C:\Users\ramra\OneDrive\Desktop\Work Projects\Kooner Fleet Management Solutions Sediment Analysis .csv"

data = pd.read_csv(df)
data.head()  
```

```python

# Combine all columns with fragmented comments into a single string
data['Combined_Comments'] = data.apply(lambda row: ' '.join(row.fillna('').astype(str)), axis=1)

# Drop empty rows
data = data[data['Combined_Comments'].str.strip() != '']

# Preview the cleaned data
data.head()
```

```python
# Combine all comments into one string
all_comments = ' '.join(data['Combined_Comments'].tolist())

# Calculate word frequencies
vectorizer = CountVectorizer(stop_words='english')
word_counts = vectorizer.fit_transform(data['Combined_Comments'])
words = vectorizer.get_feature_names_out()
frequencies = word_counts.sum(axis=0).A1

# Create a DataFrame for word frequencies
word_freq_df = pd.DataFrame({'Word': words, 'Frequency': frequencies}).sort_values(by='Frequency', ascending=False)

# Save the word frequencies to Excel
word_freq_df.to_excel('KFM Frequency.xlsx', index=False)
```

```python
# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='viridis').generate(all_comments)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Comments')
plt.show()

# Save the word cloud as an image
wordcloud.to_file('word_cloud.png')
```
